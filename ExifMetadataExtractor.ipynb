# Reading EXIF metadata using MetadataExtractor

The [MetadataExtractor](https://github.com/drewnoakes/metadata-extractor-dotnet) library allows for rapid extration of various types of metadata from many image file formats. It's licensed under Apache 2.0 license, allowing free commercial use.

```polyglot-notebook
#r "nuget: MetadataExtractor"
```

```polyglot-notebook
using MetadataExtractor;
```

```polyglot-notebook
var imageFileName = @"demo.jpg";
```

```polyglot-notebook
var directories = ImageMetadataReader.ReadMetadata(imageFileName);
directories.Display();
```
