# Reading EXIF metadata using the MetadataExtractor library

The [MetadataExtractor](https://github.com/drewnoakes/metadata-extractor-dotnet) library allows for rapid extration of various types of metadata from many image file formats. It's licensed under Apache 2.0 license, allowing free commercial use, making it a better option than `SixLabors.ImageSharp` for commercial use. 

It can only read the data and does not allow you to access raw data easily.

```polyglot-notebook
#r "nuget: MetadataExtractor"
```

Include namespace:

```polyglot-notebook
using MetadataExtractor;
```

Define file name:

```polyglot-notebook
var imageFileName = @"demo.jpg";
```

The metadata are organized by _directories_, which are basically metdata standards. We can prowse trough the tree:

```polyglot-notebook
var directories = ImageMetadataReader.ReadMetadata(imageFileName);
directories.Display();
```

If we want flat structure, we can use LINQ and its `SelectMany` method:

```polyglot-notebook
using System.Linq;

var flattened = directories.SelectMany(dir => dir.Tags.Select(tag => new {
    Directory = dir.Name,
    Tag = tag.Name,
    Value = tag.Description
}));

flattened.DisplayTable();
```
